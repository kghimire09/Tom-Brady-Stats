```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('business.xlsx')
print(df.head(10))
```

```python
mean = df['Passes Completed '].mean()
print (mean)
```

```python
median = df['Total Team Touchdowns'].median()
print(median)
```

```python
sd = df['Passes Completed '].std()
print(sd)
```

```python
x_values = df['Year']
y_values = df['Passes Completed ']
another = df['Total Team Touchdowns']
```

```python
plt.figure(figsize = (10,10))
plt.plot(x_values, y_values,color = 'green', linestyle = '--')
plt.plot(x_values, another, color = 'red', linestyle = ':')
plt.xlabel('Year')
plt.ylabel('Number of passes/touchdowns')
plt.title('Tom Brady career passes/touchdown')
plt.legend(['Passes','Touchdowns'])
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
plt.xlabel('Year')
plt.ylabel('Passes')
plt.title('Tom Brady Total Passes Each Year')
plt.plot(x_values,y_values,color = '#FF0101',linestyle = '--')
plt.subplot(1,2,2)
plt.xlabel('Year')
plt.ylabel('Touchdowns')
plt.title('Tom Brady Total Touchdowns Each Year')
plt.plot(x_values,another,color = 'green', linestyle = '--')
plt.show()
```
